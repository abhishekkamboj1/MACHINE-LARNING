```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv", encoding='unicode_escape')
#to avoid encoding error, use unicode_escope
```

```python
#data cleanning
df.info()
```

```python
#droop blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
#check the null values
df.isnull().sum()
```

```python
df.shape
```

```python
#drop null values
#inplace=True) ye permanent save kar deta hai data mai
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
#chect the data type
df['Amount'].dtype
```

```python
#EDA[Exploratory Data Analysis]
```

```python
df.columns
```

```python
#Univariak analysis
kk= sns.countplot(x='Gender', data=df, color='red', edgecolor="black")

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#observasion 
#we can see that most of the buyers are females as compare to males. 
```

```python
#Univariak analysis
kk=sns.histplot(data=df, x='Age Group', palette='viridis', edgecolor='black')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#observasion
#The most of buyers are of age group between 26-35years
```

```python
#Bivariate Analysis
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,6)})
kk=sns.barplot(data= sales_state, x='State', y='Orders', edgecolor='black', palette='viridis')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#observation
#top three orders states are Uttar Pardesh, Maharashtra and Karnataka.
```

```python
#Univariak analysis
sns.set(rc={'figure.figsize':(5,3)})
kk=sns.countplot(data=df, x='Marital_Status', edgecolor='black', color='green')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#observation
#0 is married
#1 is unmarried
#married are are ordered more then unmarried persons
```

```python
#Univariak analysis
sns.set(rc={'figure.figsize':(20,5)})
kk=sns.countplot(data=df, x='Occupation', edgecolor='black', palette='viridis')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#observation 
#most of the buyers from it sectors and healthcare.
```

```python
#Bivariate Analysis
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
kk=sns.barplot(data=sales_occupation, x='Occupation',y='Amount', edgecolor='black')
```

```python
#observasion
#the most of ammount spend it sectors and healthcare.
```

```python
#Bivariate Analysis
sns.set(rc={'figure.figsize':(25,3)})
kk=sns.countplot(data=df, x='Product_Category', edgecolor='black')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
#Bivariate Analysis
sales_pro_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)
sns.set(rc={'figure.figsize':(13,3)})
kk=sns.barplot(data=sales_pro_category, x='Product_Category',y='Amount', edgecolor='black')

plt.title("Distribution of Brands")
plt.xlabel("Brands Name")
plt.ylabel("Count")
```
