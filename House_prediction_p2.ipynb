```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# check the null value
df.isnull().sum()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date'] = df['date'].dt.year
# Check the data type of the 'date' column
print(df['date'].dtypes)
```

```python
df['price'] = df['price'].astype(int)
```

```python
# drop the extra columns 
colms = ["street", "city", "statezip", "country"]
df.drop(colms, axis=1, inplace=True)
df.columns
```

```python
df.info()
```

```python
#Bivariate Analysis
plt.scatter(df['bedrooms'], df['price'])
plt.xlabel('bedrooms')
plt.ylabel('price')
```

```python
#Bivariate Analysis
sns.barplot(data=df, x='floors', y='price')
```

```python
#Bivariate Analysis
kk=sns.barplot(data=df, x='view', y='price')
```

```python
#Univariak analysis
kk=sns.countplot(data=df, x='waterfront')

for sum in kk.containers:
    kk.bar_label(sum)
```

```python
# give the range
X=df.iloc[:, 0:12]
Y=df.iloc[:, -13]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
X_test
```

```python
Y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1, -1))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred = lr.predict(X_test)
```

```python
Y_test.values
```

```python
print("MAE", mean_absolute_error(Y_test, y_pred))
```

```python
print("MSE", mean_squared_error(Y_test, y_pred))
```

```python
print("RMSE",np.sqrt( mean_squared_error(Y_test, y_pred)))
```

```python
print("MSE",r2_score(Y_test, y_pred))
r2 = r2_score(Y_test,y_pred)
```
